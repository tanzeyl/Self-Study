```python
import matplotlib.pyplot as plt
list1 = [2,4,6,1]
plt.plot(list1)
plt.ylabel("Values")
plt.xlabel("Indices")
plt.title("My Plot")
plt.grid()
```

```python
import numpy as np
t = np.arange(0, 5, 0.2)
plt.plot(t, t**2, "b--", label = " ^2")
plt.plot(t, t**2.2, "rs", label = " ^2.2")
plt.plot(t, t**2.5, "g^", label = " ^2.5")
plt.legend()
plt.grid()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [1,4,6,8]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0], color = 'r', linewidth = 2)
plt.setp(lines[1], color = 'y', linewidth = 2)
```

```python
def f(t):
  return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0,100,2)
t2 = np.arange(0,700,3)
plt.figure(1)
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b--')
plt.subplot(212)
plt.grid()
plt.plot(t2, f(t2), 'r--')
```
