```julia
histogram(y)
```

```julia
p1 = plot(x, y, title = "Random Line Chart", label = "Line Chart for Y", xlabel = "Values of X", ylabel = "Values of Y")
p2 = scatter(x, y, title = "Scatter Plot", label = "Bubbles for Y", xlabel = "Values of X", ylabel = "Values of Y")
p3 = bar(x, y, title = "Bar Plot", xlabel = "Values of X", ylabel = "Values of Y", label = "Bar for Y")
p4 = histogram(y, title = "Hostogram", xlabel = "Values of Y", ylabel = "Frequency", label = "Value counts of Y")
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```
