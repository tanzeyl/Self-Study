```python
#Replacing missing values with means
import numpy as np
mean = data["city-mpg"].mean()
data["city-mpg"].replace(np.nan, mean)
```

```python
#Converting city-mpg to city-L/100km
data["city-mpg"] = 235/data["city-mpg"]
data.rename(columns={"city-mpg" : "city-L/100Km"}, inplace = True)
```

```python
data.head()
```

```python
#Converting "length" column from float to integer
data["length"] = data["length"].astype("int")
```

```python
data.dtypes
```

```python
#Binning the "city-L/100Km" column
bins = np.linspace(min(data["city-L/100Km"]), max(data["city-L/100Km"]), 4)
group_names = ["Low", "Medium", "High"]
data["L/100-binned"] = pd.cut(data["city-L/100Km"], bins, labels = group_names, include_lowest = True)
```

```python
#One-hot-encoding
data2 = pd.get_dummies(data["fuel-type"])
```

```python
#Use of groupby method
import matplotlib.pyplot as plt
df_test = data[["drive-wheels", "body-style", "city-L/100Km"]]
df_grp = df_test.groupby(["drive-wheels", "body-style"], as_index = False).mean()
df_pivot = df_grp.pivot(index = "drive-wheels", columns = "body-style")
plt.pcolor(df_pivot, cmap="RdBu")
plt.colorbar()
```
