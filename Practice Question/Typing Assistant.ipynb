```python
import random
import time
import matplotlib.pyplot
```

```python
words = ["sasquatch", "chlamedia", "floccinaucinihilipilification", "defenstrate", "artillery", "phytoplankton", "assassination", "armadillo", "enterpreneur", "photosynthesis"]
```

```python
i = 0
n = random.randint(0,9)
mistakes = 0
times = []
print("Type this word: {}.".format(words[n]))
while len(times) < 5:
    t1 = time.time()
    a = input()
    t2 = time.time()
    t = t2-t1
    times.append(t)
    if(a != words[n]):
        mistakes = mistakes + 1
print("You made {} mistakes.".format(mistakes))
print("Now let's see your evolution.")
time.sleep(3)
x = [1,2,3,4,5]
y = times
matplotlib.pyplot.ylabel("Words typed correctly.")
matplotlib.pyplot.xlabel("Time taken.")
matplotlib.pyplot.title("Performance.")
matplotlib.pyplot.plot(x,y)
```
